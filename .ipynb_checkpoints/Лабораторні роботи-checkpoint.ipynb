### Пошук мінімального острового дерева алгоритмом Крускала

```python
import numpy as np
from lab1.graph import Graph
import networkx as nx
import matplotlib.pyplot as plt
with open("lab1/l1_2.txt",'r') as f:
    n = int(f.readline())
    string = f.readlines()
string_matrix = list(map(lambda y: y.strip().split(' '),string))
matrix = np.array([list(map(int, x)) for x in string_matrix])
print(f"Розмірність: {n}")
print(matrix)
```

```python
vis1 = nx.Graph(matrix)
plt.figure(figsize=(16, 9))
nx.draw_circular(vis1,with_labels=True)
```

```python
g = Graph(n)
for i in range(n-1):
    for j in range(i,n):
        if matrix[i][j] > 0:
            g.addEdge(i, j, matrix[i][j])
result = g.KruskalMST()
```

```python
res = np.zeros((n,n))
for v in result:
    res[v[0]][v[1]] = v[2]
    res[v[1]][v[0]] = v[2]
res
```

```python
vis2 = nx.Graph(res)
plt.figure(figsize=(16, 9))
nx.draw_circular(vis2,with_labels=True)
```
